```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
MCMC = np.loadtxt("SampleDoubleLorenz.txt")
```

### Histograms

```python
for i in range(MCMC.shape[1]):
    plt.hist(MCMC[:,i])
    plt.title('Coefficient  {}'.format(i))
    plt.show()
```

### "Trajectories" with respect to burn-in

```python
i=0
plt.scatter(np.arange(MCMC.shape[0]),
                MCMC[:,i],
           alpha = 0.01)
plt.title('Coefficient  {}'.format(i))
plt.show()
```

```python
plt.scatter(MCMC[:,1],
           MCMC[:,7],
           alpha=0.01,
           marker = ".")
plt.show()
```
